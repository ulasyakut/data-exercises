Dropping Notes and Status columns due to amount of missing value

```python
salaries.drop(['Notes','Status'],axis=1,inplace=True)
```

```python
salaries['Agency'].unique()
```

```python
salaries.drop('Agency',axis=1,inplace=True)
```

Some columns need detailed look

```python
salaries['Benefits']
```

```python
salaries.loc[salaries['Benefits'] == 'Not Provided','Benefits'] = 0
salaries.loc[salaries['OvertimePay'] == 'Not Provided','OvertimePay'] = 0
salaries.loc[salaries['OtherPay'] == 'Not Provided','OtherPay'] = 0
```

```python
salaries.fillna({'Benefits': 0,'OtherPay': 0,'OverTimePay': 0}, inplace=True)
```

```python
salaries
```

```python
salaries['BasePay'] = pd.to_numeric(salaries['BasePay'], errors='coerce')
```

```python
salaries.drop(salaries[(salaries['BasePay'] == 0) | (salaries['BasePay'].isna())].index, inplace=True)
```

```python
salaries['Benefits'] = pd.to_numeric(salaries['Benefits'], errors='coerce')
```

```python
salaries['Benefits'].max()
```

```python
salaries
```

We have cleaned data now.What we have done so far: 
-Dropped 2 columns
-Dropped all rows where BasePay == 0
-Set value of OverTimePay,OtherPay and Benefits to 0 if they were non integer values
